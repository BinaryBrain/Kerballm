```python
parts_cfg = dict()
```

```python
def list_files(my_root_dir):
    for path, _, files in os.walk(my_root_dir):
        yield path, files
```

```python
for path, files in list_files('/Applications/KSP_osx/GameData/Squad/Parts'):
    for file in files:
        if file.split('.')[-1] == 'cfg':
            key = file.split('.')[0]
            f = open(f"{path}/{file}", "r")
            rd = f.read()
            name_v = rd.split('name = ')[1].split('\n')[0]
            if key != parts_cfg.keys():
                parts_cfg[key] = name_v
```

```python
with open("parts_cfg.json", "w") as outfile: 
    json.dump(parts_cfg, outfile)
```

```python
for pn in list(parts_cfg.values()):
    with open(f"parts/{pn}", "w") as file:
        pn_text = f'part = {pn}'.join(text.split('part = '))
        file.write(pn_text[1:])
```
